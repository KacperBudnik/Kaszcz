```julia
X=file[:,:Kol];
```

```julia
q1,q3=quantile(X,0.25),quantile(X,0.75)
```

```julia
iqr=q3-q1
```

```julia
q1-1.5*iqr,q3+1.5*iqr
```

```julia
histogram(X[(X.<q3+1.5*iqr).&(X.>q1-1.5*iqr)], bins=30)
```

```julia
X[(X.>=q3+1.5*iqr).|(X.<=q1-1.5*iqr)]
```
