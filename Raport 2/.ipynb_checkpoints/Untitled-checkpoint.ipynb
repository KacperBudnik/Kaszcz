# Micro

```julia
File=CSV.read("Microsoft_Stock.csv", DataFrame);
```

```julia
N=length(File[:,:Date])
```

```julia
file=File[:,[:Close]]
file[:,:time]=[i for i in 1:N];
```

```julia
file
X=file[:,:Close];
```

```julia
γ(X,h)=begin
    n=length(X)
    x=mean(X)
    sum(@. (X[h+1:end]-x)*(X[1:n-h]-x) )/n
end
```

```julia
γ₀=γ(X,0)
γ₁=γ(X,1)
Γ=[γ₀ γ₁;
   γ₁ γ₀]
γ₂=[γ₁, γ(X,2)];
```

```julia
Φ=inv(Γ)*γ₂
```

```julia
A(p)=[γ(X,abs(i-j)) for i in 1:p, j in 1:p]
```

```julia
A(2)
```

```julia
b(p)=[γ(X,i) for i in 1:p]
```

```julia
φ(p)=inv(A(p))*b(p)
```

```julia
σ(p)=begin
    a=φ(p)
    c=b(p)
    γ(X,0)-sum(a[i]*c[i] for i in 1:p)
end
```

```julia
σ(200)
```

```julia
FPE=[σ(p)*(N+p)/(N-p) for p in 1:100]
```

```julia
(1:100)[argmin(FPE)]
```

```julia
φ(2)
```
