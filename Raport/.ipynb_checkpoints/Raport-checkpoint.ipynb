# Wczytanie danych i odrzucenie odstających

```julia
file=CSV.read("diamonds.csv", DataFrame);
```

```julia
file[:,:V]=file[:,:x] .* file[:,:y] .* file[:,:z];
```

```julia
plot(boxplot(file[:, :V]),boxplot(file[:, :price]))
```

```julia
qpₘᵢₙ,qpₘₐₓ=quantile(file[:,:price], 0.01),quantile(file[:,:price],0.90);
```

```julia
qvₘᵢₙ,qvₘₐₓ=quantile(file[:,:V], 0.01),quantile(file[:,:V], 0.95);
```

```julia
res=file[((file[:, :V].<qvₘₐₓ) .& (file[:, :V].>qvₘᵢₙ)) .& ((file[:, :price].<qpₘₐₓ) .& (file[:, :price].>qpₘᵢₙ)), [:V,:price]];
```

```julia
size(file), size(res)
```

```julia
(size(file)[1] -size(res)[1])/size(file)[1]
```

```julia
plot(boxplot(res[:, :V]),boxplot(res[:, :price]))
```

## Dopasowanie mnodelu

```julia
@. model(x,p)=p[1]+p[2]*x^p[3]
@. modell(x,p)=p[1]+p[2]*x
```

```julia
fit = curve_fit(model, res[:,:V], res[:,:price], [0.,1e2,1.2])
fitl = curve_fit(modell, res[:,:V], res[:,:price], [0.,1e2]);
```

```julia
println(fit.param)
println(fitl.param)
```

```julia
scatter(res[:,:V],res[:,:price])
plot!(sort(res[:,:V]), t->model(t,fit.param));
```
