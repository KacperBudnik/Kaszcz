```julia
b(β,σ)=begin
    Y=X*β.+σ.*[sqrt(-2log(rand()))*cos(2pi*rand()) for i in 1:n];
    x=mean(X)
    y=mean(Y)
    sum((X.-x) .* (Y.-y))/sum( (X.-x).^2 )
end
```

```julia
B=[b(β,σ) for i in 1:M];
```

```julia
mean(B), var(B)
```

```julia
E=[]
V=[]
for σ in 0.1:0.1:3
    B=[b(β,σ) for i in 1:M];
    append!(E,mean(B))
    append!(V, var(B))
end
```

```julia
scatter(0.1:0.1:3,E)
```

```julia
scatter(0.1:0.1:3,V)
```

## Wariant 2

```julia
b(β,σ,n)=begin
    X=0:10/(n-1):10
    Y=X*β.+σ.*[sqrt(-2log(rand()))*cos(2pi*rand()) for i in 1:n];
    x=mean(X)
    y=mean(Y)
    sum((X.-x) .* (Y.-y))/sum( (X.-x).^2 )
end
```

```julia
E=[]
V=[]
σ=1
for n in 10:10:2000
    B=[b(β,σ,n) for i in 1:M];
    append!(E,mean(B))
    append!(V, var(B))
end
```
