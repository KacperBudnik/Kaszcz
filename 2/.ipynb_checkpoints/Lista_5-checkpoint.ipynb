```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss
```

```python
sig_w = 1
sig_z = 2
n = 1000
phi = 0.5

Z = np.random.normal(0, sig_z, n+1)
S = np.random.normal(0, sig_w, n)
X = Z[1:]+phi*Z[:-1] + S
```

```python
def autokow(x, h):
    n = len(x)
    xm = np.mean(x)
    return 1/n*np.sum((x[h:]-xm)*(x[:n-h]-xm))
```

```python
H = np.arange(0, 100, 1)
kow = []
for h in H:
    kow.append(autokow(X, h))
plt.scatter(H, kow)

te = np.zeros(len(H))
te[0] = sig_w**2+sig_z**2*(1+phi**2)
te[1] = phi*sig_z**2

plt.plot(H, te, color="orange")
plt.show()
```
